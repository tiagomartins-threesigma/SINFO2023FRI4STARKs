The following code performs the sequence of operations:
1. defines $c$, the array of polynomial coefficients of $p$,
2. computes the evaluation of $p$ by computing the $\texttt{FFT}$ of the array of coefficients,
3. prints the previous two results,  
-----
4. enlarges the array of polynomial coeffients by a factor of $\mathrm{BUF}\times$ by appending zeros,
5. computes the more points in the evaluation of $p$ by computing the $\texttt{FFT}$ of the enlarged array of coefficients,
6. prints the previous two results.
